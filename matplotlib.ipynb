```python
# bar graph 

x = ["python", "java", "c", "c++"]
y = [90, 85, 60, 65]
z = [20,40,35,55]
colors = ["y", "b", 'm', "g"]
plt.xlabel("languages",fontsize=10)
plt.ylabel("No", fontsize=10)
plt.title("Ashutoshcodes",fontsize=10)
plt.bar(x,y, width=0.3,linewidth=3,edgecolor="y",label="popularity")
plt.bar(x,z, width=0.3, linewidth=3, edgecolor="r")
plt.legend()
plt.show()
     
```

```python
# scatter plot
import matplotlib.pyplot as plt
days = [1,2,3,4,5]
no = [1,3,5,7,9]
colors = ["y", "g", "b", "r", "m"]
sizes = [400,300,500,600,800]
plt.scatter(days,no,c=colors, s=sizes, marker="*", alpha=0.3, linewidth=4, edgecolor="r")
plt.title("Scattersplot", fontsize=8)
plt.xlabel(days, fontsize=10)
plt.ylabel(no, fontsize=10)
plt.show()
```

```python
days = [1,2,3,4,5]
no = [1,3,5,7,9]
colors = [10,34,56,78,45]
sizes = [400,300,500,600,800]
plt.scatter(days,no,c=colors, s=sizes, cmap="viridis")
t = plt.colorbar()
t.set_label("colorbar", fontsize=15)
plt.title("Scattersplot", fontsize=8)
plt.xlabel(days, fontsize=10)
plt.ylabel(no, fontsize=10)
plt.show()
```

```python
# stem plot

x = [5,2,6,7,1]
y = [3,2,1,5,5]

plt.stem(x,y, linefmt= ":", markerfmt= "r+", basefmt= "y", label="python", orientation="vertical")
plt.legend()
plt.show()
```

```python
#multiple box plots
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70]
x1 = [10,20,30,40,50,60,70,90]
y = [x,x1]

plt.boxplot(y, notch=False, tick_labels=["python", "c++"], patch_artist=False, showmeans=True)
           
plt.show()
```

```python
# stack plot

x = [1,3,5,7,9]
area1 = [2,4,6,8,10]
area2 = [1,2,3,4,5]
area3 = [2,3,5,7,6]
l = [x, area1, area2, area3]
plt.stackplot(x,area1,area2,area3, labels=l, colors=["r","g","y"])
plt.title("Python")
plt.xlabel("data")
plt.ylabel("no.")
plt.legend(loc=2)

plt.show()
```

```python
# step plot
x = [1,3,5,7,9]
y = [2,4,6,8,9]

plt.step(x, y, color="y", marker="o", ms=10, label="python")
plt.legend()
plt.xlabel(" x - axis")
plt.ylabel(" y - axis")
plt.grid()
plt.show()
```

```python
# fill between plot
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

plt.plot(x, y, color="r", label="python")
plt.legend()
plt.fill_between(x, y, where=(x>=2)&(x<=4))
plt.xlabel(" x - axis")
plt.ylabel(" y - axis")
plt.show()
```

```python
# sub plot in matplotlib
x = [1,2,3,4,5]
y = [2,3,5,7,9]
plt.subplot(2,2,1)
plt.plot(x, y, color="g")

plt.subplot(2,2,2)
plt.pie([1])

x1 = [1,3,5,7,9]
plt.subplot(2,2,3)
plt.pie(x1)

x2= [3,5,7,4,8]
y1 = [4,5,6,7,9]
plt.subplot(2,2,4)
plt.bar(x2, y1)


plt.show()
```

```python
# axis in matplotlib

x = [1,2,3,4]
y = [1,2,4,3]

plt.plot(x , y)
plt.xticks(x, labels=["python", "c", "c++", "java"])
plt.yticks(x)

plt.axis([0,10,0,7])
plt.show()
```

```python
# text in matplotlib

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)
plt.xlabel("days")
plt.ylabel("works")
plt.title("Python",fontsize=15)
plt.text(2,5, "java", fontsize=15, style="italic", bbox={"facecolor":"r"})
plt.annotate("python", xy=(1,3), xytext=(4,5), arrowprops=dict(facecolor="black", shrink=100))
plt.show()

```
